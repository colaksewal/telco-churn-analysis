

1. **`dtypes` nedir?**
   - `dtypes`, DataFrame'in her kolonunun veri tipini döner. Örneğin, kolonlarınızda integer, float veya object (string) veri tipleri olabilir. `dtypes`, bu veri tiplerini bir Pandas Series olarak döner.

2. **`pipe` nedir?**
   - `pipe` metodu, DataFrame'de bir veya daha fazla fonksiyonu zincirleme olarak uygulamak için kullanılır. Fonksiyonlar, veriyi işlemden geçirerek ardışık bir şekilde uygulanır. Bu, daha okunabilir ve organize bir kod yazmanıza yardımcı 
```python
import pandas as pd

# Örnek DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)

# Fonksiyonlar
def add_two(df):
    df['A'] = df['A'] + 2
    return df

def multiply_by_five(df):
    df['B'] = df['B'] * 5
    return df

# pipe ile kullanımı
df = df.pipe(add_two).pipe(multiply_by_five)
print(df)

```


olur.

3. **`factorize` nedir?**
   - `factorize`, kategorik bir kolonun benzersiz değerlerini sayısal kodlara dönüştürür. Örneğin, ["kırmızı", "mavi", "yeşil"] gibi kategorik değerler 0, 1, 2 gibi sayısal değerlere dönüştürülür. Bu, kategorik verileri sayısal olarak işlemek veya modellemek için kullanışlıdır.

Şimdi bu fonksiyonların bir araya geldiği kodu inceleyelim:

```python
# DataFrame'in kolonlarının veri tiplerini alır ve 'Object' veri tipine sahip kolonların isimlerini alır
char_cols = df.dtypes.pipe(lambda x: x[x == 'Object']).index

# Kategorik kolonları sayısal kodlara dönüştürür
for c in char_cols:
    df[c] = pd.factorize(df[c])[0]

# İlk birkaç satırı görüntülemek için
df.head()
```

**Kod Açıklaması:**
1. `df.dtypes.pipe(lambda x: x[x == 'Object']).index`: İlk olarak, DataFrame'in tüm kolonlarının veri tiplerini alır (`df.dtypes`). `pipe` metodu, bu veri tiplerine anonim bir fonksiyon (lambda) uygulayarak sadece `Object` veri tipine sahip kolonları seçer. Son olarak, bu kolonların isimlerini (`index`) alır ve `char_cols` değişkenine atar.
  
2. `for c in char_cols: df[c] = pd.factorize(df[c])[0]`: Daha sonra, `char_cols` içindeki her kolon için `factorize` fonksiyonunu kullanarak kategorik değerleri sayısal değe

   lere dönüştürür. `factorize` iki değer döner:
   birinci değer dönüşen sayısal kodları içerir, iki

   ci değer ise benzersiz kategorileri listeler. Biz sadece birinci değeri alırız (`[0]`) ve orijinal kolonu bu yeni sayısal değerlerle güncelleriz.

Bu şekilde, tüm kategorik (object tipindeki) khtiyacınız olursa, bana bildirmekten çekinmeyin!olursa, lütfen bana bildirin!

````python
char_cols = df.dtypes.pipe(lambda x: x[x == 'object']).index

for c in char_cols:
    df[c] = pd.factorize(df[c])[0]

df.head()
````

df1.groupby(x)[y]: df1 DataFrame'ini x değişkenine göre gruplar ve y (bu, Churn değişkeni olabilir) değişkenine göre alt gruplar oluşturur.

.value_counts(normalize=True): Her gruptaki y değişkeninin frekanslarını hesaplar ve bunları normalize ederek yüzdelerini bulur.

.mul(100): Yüzdeleri 100 ile çarparak yüzdelik değerlere dönüştürür.

.rename('percent'): Hesaplanan yüzdeleri percent olarak yeniden adlandırır.

.reset_index(): Gruplama işlemi sonrası oluşan çok seviyeli indeksleri (multi-index) resetler, böylece düz bir DataFrame elde edilir.

````python
import seaborn as sns
````

````python
y = 'Churn'

for x in catvars:
    plot = df1.groupby(x)[y]\
        .value_counts(normalize=True).mul(100)\
        .rename('percent').reset_index()\
        .pipe((sns.catplot,'data'), x=x, y='percent', hue=y, kind='bar',palette=["hotpink", "#87ceeb"], height = 16, aspect = 1.5)
    plot.fig.suptitle("Churn by " + x)
    plot
````

````python

# Churn müşterilerinin sayısını al
churners_number = len(df[df['Churn'] == 1])
print("Number of churners", churners_number)

# Churn müşterilerini seç
churners = df[df['Churn'] == 1]

# Churn olmayan müşterilerden rastgele örnek al
non_churners = df[df['Churn'] == 0].sample(n=churners_number)
print("Number of non-churners", len(non_churners))

# Churn ve churn olmayan müşterileri birleştir
df3 = pd.concat([churners, non_churners])

````

````python
def show_correlations(df, show_chart = True):
    fig = plt.figure(figsize = (30,10))
    corr = df.corr()
    if show_chart == True:
        sns.heatmap(corr, 
                    xticklabels=corr.columns.values,
                    yticklabels=corr.columns.values,
                    annot=True)
    return corr
````

````python
correlation_df = show_correlations(df3,show_chart=True)
````

````python
# Define the target variable (dependent variable) 
y = df.Churn 
df = df.drop(['Churn'], axis= 1)
````

````python
df
````

````python
# Splitting training and testing data
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.20)  
````

````python
# Applying Support Vector Machine algorithm
from sklearn.svm import SVC  
svclassifier = SVC(kernel='linear', degree=8)  
svclassifier.fit(X_train, y_train)
````

````python
# Predicting part, applying the model to predict
y_pred = svclassifier.predict(X_test) 
````

````python
# Evaluating model performance
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))
````
